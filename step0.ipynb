```python
import glob, random
import pandas as pd
import numpy as np

files = glob.glob(r"D:\MERGER IMPACT ANALYSIS\t100_raw\*.asc")
sample_files = random.sample(files, 10)

SEP="|"
ORIG_IDX=3
DEST_IDX=7

def uniq_pairs(path, nrows=20000):
    x = pd.read_csv(path, sep=SEP, header=None, engine="python", nrows=nrows, usecols=[ORIG_IDX, DEST_IDX])
    o = pd.to_numeric(x[ORIG_IDX], errors="coerce")
    d = pd.to_numeric(x[DEST_IDX], errors="coerce")
    ok = (~o.isna()) & (~d.isna())
    o = o[ok].astype(int).to_numpy()
    d = d[ok].astype(int).to_numpy()
    lo = np.minimum(o,d)
    hi = np.maximum(o,d)
    return len(set(zip(lo,hi))), ok.mean()

for fp in sample_files:
    u, share_ok = uniq_pairs(fp)
    print(fp.split("\\")[-1], "uniq_pairs:", u, "ok_share:", round(share_ok,3))
```

```python
import re

raw = pd.read_csv("t100_raw/2010_Jan.asc", sep="|", header=None, engine="python")

def looks_like_airline(x):
    # Common airline code patterns in BTS: 2 letters (AA, US, DL, WN) sometimes 3 alnum (e.g., "3M" etc.)
    s = str(x).strip()
    return bool(re.fullmatch(r"[A-Z0-9]{2,3}", s))

candidates = []
for j in range(raw.shape[1]):
    col = raw[j].astype(str).str.strip()
    # ignore purely numeric columns quickly
    if col.str.fullmatch(r"\d+(\.\d+)?").mean() > 0.98:
        continue
    share = col.apply(looks_like_airline).mean()
    top = col.value_counts().head(10).index.tolist()
    if share > 0.50:  # threshold: lots of values look like airline codes
        candidates.append((j, share, top))

sorted(candidates, key=lambda x: -x[1])[:10]
```

```python
import pandas as pd
import numpy as np

raw_path = r"D:\MERGER IMPACT ANALYSIS\t100_raw\2010_Jan.asc"
raw = pd.read_csv(raw_path, sep="|", header=None, engine="python")

IDX_YEAR=0; IDX_MONTH=1; IDX_CARRIER=10
IDX_PAX=16; IDX_DEP=17

candidates = [19, 26]  # add more if you want, e.g. [18,19,20,25,26]
out = []

def to_num(s):
    return pd.to_numeric(s, errors="coerce")

for sidx in candidates:
    df = raw[[IDX_YEAR, IDX_MONTH, IDX_CARRIER, IDX_PAX, IDX_DEP, sidx]].copy()
    df.columns = ["YEAR","MONTH","carrier","pax","dep","seats_cand"]
    df["pax"] = to_num(df["pax"]).fillna(0)
    df["dep"] = to_num(df["dep"]).fillna(0)
    df["seats_cand"] = to_num(df["seats_cand"])

    # route-level keys are messy here; we’re just testing mechanical sanity overall
    bad_zero = ((df["pax"] > 0) & (df["seats_cand"] == 0)).mean()
    bad_nan  = df["seats_cand"].isna().mean()

    # load factor-ish (avoid divide by zero)
    lf = (df["pax"] / df["seats_cand"]).replace([np.inf, -np.inf], np.nan)
    lf_valid = lf.dropna()
    lf_over2 = (lf_valid > 2).mean() if len(lf_valid) else np.nan

    corr_dep = df[["dep","seats_cand"]].dropna().corr().iloc[0,1]

    out.append((sidx, bad_zero, bad_nan, lf_over2, corr_dep, df["seats_cand"].median(skipna=True)))

pd.DataFrame(out, columns=["seats_idx","share(pax>0 & seats=0)","share(seats NaN)","share(lf>2)","corr(dep,seats)","median(seats)"]).sort_values("share(pax>0 & seats=0)")
```

## Data Cleaning

```python
import pandas as pd
import numpy as np

RAW_PATH = r"D:\MERGER IMPACT ANALYSIS\t100_processed\t100_citypair_quarter.parquet"
CLEAN_PATH = r"D:\MERGER IMPACT ANALYSIS\t100_processed\t100_citypair_quarter_clean.parquet"

df = pd.read_parquet(RAW_PATH)
```

```python
df_clean = df.copy()

bad_rows = (
    (df_clean["seats"].isna()) |
    ((df_clean["seats"] == 0) & ((df_clean["pax"] > 0) | (df_clean["dep"] > 0)))
)

print("Rows before:", len(df_clean))
print("Dropping rows:", bad_rows.sum())

df_clean = df_clean.loc[~bad_rows].copy()

print("Rows after:", len(df_clean))
```

```python
# No impossible combinations
assert not ((df_clean["pax"] > 0) & (df_clean["seats"] == 0)).any()
assert not ((df_clean["dep"] > 0) & (df_clean["seats"] == 0)).any()

# Seats non-negative
assert (df_clean["seats"] >= 0).all()

# Market-time-carrier uniqueness
dup = df_clean.duplicated(["mkt","YEAR","quarter","carrier"]).mean()
print("Duplicate share:", dup)
```

```python
df_clean.to_parquet(CLEAN_PATH, index=False)
print("Saved clean parquet to:", CLEAN_PATH)
```
